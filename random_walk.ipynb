```
import numpy as np
import matplotlib.pyplot as plt
del_X = np.random.normal(0,1,(2,5))
plt.plot(del_X[0],del_X[1],"go")
X = np.cumsum(del_X, axis = 1, )
plt.plot(X[0],X[1], "ro-")
plt.savefig("rw.pdf")
```
